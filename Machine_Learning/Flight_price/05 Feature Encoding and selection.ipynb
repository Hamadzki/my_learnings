### 1.Airline

airline has 11 categories so we can't make such more dummies.

so we will use threshold other method

```python
df["Airline"].value_counts()
```

```python
def count_percentage(col):
    return df[col].value_counts() * 100 /len(df) 
```

```python
threshold = count_percentage("Airline")
```

less than 5% values_counts we will convert those into others category

```python
other_values_names = threshold[threshold<5].index
other_values_names
```

```python
df["Airline"].replace(other_values_names,"others",inplace = True)
```

**we have to also done this work in cleaned_df**

```python
def get_dummies(col):
    return pd.get_dummies(df[col],drop_first = True)
```

```python
Airline = get_dummies('Airline')
```

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.feature_selection import SelectKBest
```

```python
from sklearn.feature_selection import f_classif
from sklearn.feature_selection import mutual_info_classif
```

```python
len(x.columns)
```

```python
fs =  SelectKBest(score_func = f_classif , k = 28)
```

```python
fs.fit(x,y)
```

```python
score_df = pd.DataFrame(data = fs.scores_ ,index =  x.columns , columns = ["Score"])
score_df.sort_values(by = "Score" , ascending = False)
```

```python
selected_feature_columns = x.columns[fs.get_support()]
selected_feature_columns
```

```python
featured_encoded_df = pd.concat([encoded_df[selected_feature_columns] , encoded_df["Price"]] , axis = 1)
```

```python
featured_encoded_df.to_csv('featured_encoded_df.csv',index = False)
```
