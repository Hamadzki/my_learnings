```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
x = df.drop('Price',axis = 1)
y = df["Price"]
```

### chi

```python
model_chi = SelectKBest(score_func = chi2 , k =5)
```

```python
chi = model_chi.fit(x,y)
```

```python
#fs.scores_
```

```python
def score_df(model_fit):
    z = pd.DataFrame({"columns": x.columns,"scores":model_fit.scores_})
    z = z.sort_values(by = 'scores',ascending = False)
    z.reset_index(drop = True,inplace = True)
    return z
```

```python
score_df_chi = score_df(chi)
```

```python
score_df_chi
```
