```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
class Ann(nn.Module):
    def __init__(self,input_features = X_train.shape[1] , hidden1 = 200 , hidden2 = 100 , output = 1):
        super().__init__()
        self.layer1 = nn.Linear(input_features , hidden1)
        self.layer2 = nn.Linear(hidden1 , hidden2)
        self.output_l = nn.Linear(hidden2 , output)
    
    def forward(self , x):
        x = self.layer1(x)
        x = F.relu(x)
        
        x = self.layer2(x)
        x = F.relu(x)
        
        x = self.output_l(x)
        
        return x
```

```python
model = Ann()
```

```python
loss_function = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model.parameters() , lr = 0.01)
```
